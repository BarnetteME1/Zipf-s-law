```python
import re
import requests
```

```python
res = requests.get("https://www.gutenberg.org/files/832/832-h/832-h.htm")
robin_hood = res.text
```

```python
robin_hood = re.sub(r'\n', ' ', robin_hood)
robin_hood = re.sub(r'<.*?>', "", robin_hood)
robin_hood = re.sub(r'CHAPTER \w+', '', robin_hood)
robin_hood = robin_hood.lower()
```
